# Get validation set

```python
import sys
sys.path.append('coco-text')
import coco_text
COCO_DATA='/home/sravya/data/muse/coco/coco2017/'
ct = coco_text.COCO_Text(COCO_DATA+ 'COCO_Text.json')
```

```python
valid_img_ids = ct.getImgIds(imgIds=ct.val)
```

```python
valid_imgs = ct.loadImgs(valid_img_ids)
```

```python
valid_imgs[0]
```

# Run inference and write to JSON

```python
#TODO: Do batch inference?
import detection
import recognition

def infer(file_path):
    wordBB = None
    score = None
    text = None
    wordBB, score = detection.detect(file_path)

    if score.shape[0] == 0:
        wordBB = None
        score = None
    else:
        text = recognition.recognize(file_path, wordBB)
    return text, wordBB
```

```python
import json
from tqdm import tqdm
skipped_files = []

entries = []
for img in tqdm(valid_imgs):
    filename = img['file_name'].split('_')[2]
    try:
        texts, bboxes = infer(COCO_DATA + 'train2017/' + filename)
    except:
        skipped_files.append(filename)
    if bboxes is not None:
        for index, bbox in enumerate(bboxes):
            entry = {}
            entry['utf8_string'] = texts[index]
            entry['image_id'] = img['id']
            entry['bbox'] = [bbox[0], bbox[1], bbox[2]-bbox[0], bbox[3]-bbox[1]]
            #score
            #category
            entries.append(entry)
with open('result.json', 'w') as fp:
    json.dump(entries, fp)
```
